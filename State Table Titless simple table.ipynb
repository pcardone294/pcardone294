```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://worldpopulationreview.com/states'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[0]
```

```python
soup.find('table', class_ = 'tp-table-body is-narrow w-full min-w-full table-auto border-separate border-spacing-0 border bg-white')
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
soup.find_all('th')
```

```python
state_titles = table.find_all('th')
```

```python
state_table_titles = [title.text.strip() for title in state_titles]

print(state_table_titles)

state_headers = state_table_titles
```

```python
state_headers = state_headers[1:9]

print(state_headers)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = state_headers)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
#In the following code, we are converting the columns to their proper data types so we can make calculations off of it. In order to do this we must first remove the unneccessary characters in the strings
df['State'] = df['State'].astype(str)
df['2023 Population'] = df['2023 Population'].astype(str).str.replace(',','').astype(int)
df['Growth Rate'] = df['Growth Rate'].astype(str).str.replace('%','').astype(float)
df['2022 Population'] = df['2022 Population'].astype(str).str.replace(',','').astype(int)
df['2010 Population'] = df['2010 Population'].astype(str).str.replace(',','').astype(int)
df['Growth Since 2010'] = df['Growth Since 2010'].astype(str).str.replace('%','').astype(float)
df['% of US'] = df['% of US'].astype(str).str.replace('%','').astype(float)
df['Density (/miÂ²)'] = df['Density (/miÂ²)'].astype(str).str.replace(',', '').astype(int)
```

```python
df.head()
```

```python
print(df.info())
```

```python
df['Positive or Negative'] = df['Growth Rate'].apply(lambda x: 'Positive' if x > 0 else 'Negative')
```

```python
import numpy as np
```

```python
percentile_25 = np.percentile(df['2023 Population'], 25)
percentile_65 = np.percentile(df['2023 Population'], 65)

df['Size Category'] = 'Large'  # Default to 'Large'
df.loc[df['2023 Population'] <= percentile_25, 'Size Category'] = 'Small'
df.loc[(df['2023 Population'] > percentile_25) & (df['2023 Population'] <= percentile_65), 'Size Category'] = 'Medium'
```

```python
product = df.groupby(['Size Category'])['2023 Population'].sum().sort_values(ascending = False)
print(product)
```

```python
large_states = df[df['Size Category'] == 'Large']['State'].tolist()
print("States with 'Large' size category:", large_states)
```

```python
#Converting Numbers in the df to a percent -- move the decimal point over
decimal_places_to_move = 2
df['Growth Rate'] = df['Growth Rate'] / 10**decimal_places_to_move
df['Growth Since 2010'] = df['Growth Since 2010'] / 10**decimal_places_to_move
df['% of US'] = df['% of US'] / 10**decimal_places_to_move
```

```python
Growth = df['Growth Rate'].sum()
Population = df['2023 Population'].sum()
Projection = (Growth + 1) * Population
print(Projection)

print(df['2023 Population'].sum())
print(Growth)
```

```python
Growth_Rate = 1 + df['Growth Rate']
df['2024 Projection'] = df['2023 Population'] * Growth_Rate
df.head()
```

```python
df = df.sort_values(by='Growth Rate', ascending=False)
fig = px.bar(df, x='State', y='Growth Rate',
             labels={'2023 Population': 'Population'},
             title='States With the Fastest Growing Population',
             text='2023 Population')

fig.update_traces(hoverinfo='text', marker_color='lightskyblue')

fig.update_layout(xaxis=dict(tickangle=80), showlegend=False)

fig.show()

plt.savefig('States With the Fastest Growing Population')
```
